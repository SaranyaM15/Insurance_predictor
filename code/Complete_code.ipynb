# Implementation 4: Train-Test Split & Model Training

### Split the dataset into a training set and a testing set. 

```python
# your code here 

from sklearn.model_selection import train_test_split
# medicalcharges is the goal variable, so for the input features, take all the features except the medicalcharges feature.
Input_01 = insu.drop('medicalcharges', axis=1)  

# For target variable or output variable taking only medicalcharges variable
Output_01 = insu['medicalcharges']
```

```python
# Divide the dataset into training set and evaluation set  (model fitting = 80%, evaluating = 20%)

ner_inp, par_inp, ner_out, par_out = train_test_split(Input_01, Output_01, test_size=0.2, random_state=42)

# Showing the shapes of the resultant datasets.

print(f"input training size: {ner_inp.shape}")
print(f"input testing size: {par_inp.shape}")
print(f"output training size: {ner_out.shape}")
print(f"input testing size: {par_out.shape}")

```

### Train 2 Machine Learning Models



Select any two machine learning algorithms from Logistic Regression, Random Forest, KNN, SVM, Neural Networks, DBSCAN, or even fine-tune a pre-trained model.


```python
# your code here 
from sklearn.neighbors import KNeighborsRegressor

# K-Nearest Neighbors Regressor
model_Knear_01 = KNeighborsRegressor(n_neighbors=3)
model_Knear_01.fit(ner_inp, ner_out)
y_pred_knn = model_Knear_01.predict(par_inp)
# printing testing results 
print(y_pred_knn)
```

```python
# Support Vector Regressor (SVR)
from sklearn.svm import SVR

model_supp_ve_m_01 = SVR(kernel='rbf')
model_supp_ve_m_01.fit(ner_inp, ner_out)
y_pred_svm =model_supp_ve_m_01.predict(par_inp) 
# printing resting results          
print(y_pred_svm)
```

#### Since these are for regression models we can calculate  mean squared error metric for both of the models 
* Mean Squared Error (MSE) examines the average squared difference between predicted and actual data to assess model accuracy.


```python
from sklearn.metrics import mean_squared_error
# Calculate Mean Squared Error for both models
mean_sqr_err_knn_01 = mean_squared_error(par_out, y_pred_knn)
mse_supportvect_meac_01 = mean_squared_error(par_out, y_pred_svm)

# Display the results
print(f"KNN metric mean square error: {mean_sqr_err_knn_01}")
print(f"SVM metric mean square error: {mse_supportvect_meac_01}")
```

# Implementation 5: Cross-Validation and Model Evaluation

### Evaluate Your Models



Calculate precision, recall, F1-score, and accuracy for both models.

Since K-Nearest Neighbors and Support Vector Machine are being used for regression tasks in this context, the appropriate metrics to evaluate their performance are:

The Mean Absolute Error (MAE) is the average of the absolute differences between planned and actual data.
The Mean Squared Error (MSE) is the average of the squared differences between expected and actual values.
Root Mean Squared Error (RMSE) is the square root of mean squared error, and it measures error in the same units as the target variable.
R squared (r): A statistical metric that indicates how much variance in a dependent variable is explained by independent variables.


```python
# your code here
# Calculating metrics for KNN Regression
from sklearn.metrics import mean_absolute_error,mean_squared_error,r2_score
import numpy as np 

mean_abr_err_knn_01 = mean_absolute_error(par_out, y_pred_knn)
mean_sqr_err_knn_01 = mean_squared_error(par_out, y_pred_knn)
root_mean_sqr_err_knn_01 = np.sqrt(mean_sqr_err_knn_01)
ro22_scroe_knn_01= r2_score(par_out, y_pred_knn)

# Displaying the KNN results
print("KNN Metrics:")
print(f"mean absolute error : {mean_abr_err_knn_01}")
print(f"mean square error: {mean_sqr_err_knn_01 }")
print(f"root mean Square error : {root_mean_sqr_err_knn_01}")
print(f"r square : {ro22_scroe_knn_01}")
```

```python
# Calculating metrics for SVM Regression

# Calculate metrics for SVM
mean_abr_err_svm_01 = mean_absolute_error(par_out, y_pred_svm)
mean_sqr_err_svm_01 = mean_squared_error(par_out, y_pred_svm)
roote_mean_sqr_svm_01 = np.sqrt(mean_sqr_err_svm_01)
r002_svm_01 = r2_score(par_out, y_pred_svm)

#Displaying the SVM metrics 
print("\nSVM Metrics:")
print(f"meann abs err : {mean_abr_err_svm_01}")
print(f"meann Squrt err : {mean_sqr_err_svm_01 }")
print(f"root mean square error:{roote_mean_sqr_svm_01}")
print(f"r squared score: {r002_svm_01 }")

```

### Perform Cross-Validation

Use cross-validation to check the stability of your models across different subsets of the data.

```python
# your code here
from sklearn.model_selection import cross_val_score

# Perform cross-validation
cross_vald_scr_kNear = cross_val_score(model_Knear_01, Input_01, Output_01, cv=5, scoring='neg_mean_squared_error')
crss_vald_sc_sup_vec= cross_val_score(model_supp_ve_m_01, Input_01, Output_01, cv=5, scoring='neg_mean_squared_error')

# Convert negative MSE to positive for interpretation
cros_vl_sco_01 = -cross_vald_scr_kNear
cros_vl_sco_02 = -crss_vald_sc_sup_vec

# Calculate average and standard deviation of cross-validation scores
knn_mean_mse = cros_vl_sco_01.mean()
knn_std_mse = cros_vl_sco_01.std()

svm_mean_mse = cros_vl_sco_02.mean()
svm_std_mse = cros_vl_sco_02.std()

# Display the results
print("K Nearest Neighbours  Validation Results:")
print(f"Mean MSE: {knn_mean_mse:.2f}  {knn_std_mse:.2f}")

print("\nSVM Cross-Validation Results:")
print(f"Mean MSE: {svm_mean_mse:.2f}  {svm_std_mse:.2f}")
```
